```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

df  = pd.read_csv("classpass.csv")
#df.tags = df.tags.fillna("unknown")
df.sample(15)
```

```python
df.livestream.value_counts()
```

```python
lis_tags = df.tags.dropna()
lis_tags = lis_tags.apply(eval)
```

```python

def to_1D(series):
 return pd.Series([x for _list in series for x in _list])

taglist = to_1D(lis_tags).value_counts()
top20tag = taglist[0:20]
top20tag
```

```python
#Funnel chart for top 20 tags
fig = px.funnel(top20tag, x=top20tag.values, y=top20tag.index,
               labels=dict(index="Tag"))
fig.show()
```
